```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
metrics = pd.read_csv('metrics.csv', index_col='Unnamed: 0')
metrics.head()
```

```python
fig, ax = plt.subplots(figsize=(20, 5))

sns.barplot(data=metrics, x=metrics.index, y='PER', ax=ax, color='skyblue', label='PER')
sns.lineplot(data=metrics, x=metrics.index, y='RB', ax=ax, color='red', label='Rentabilidad Bruta')
sns.lineplot(data=metrics, x=metrics.index, y='difference', ax=ax, color='green', label='Difference')

ax.set_xlabel('Distrito')
ax.set_ylabel('PER', color='skyblue')
ax.set_ylabel('Rentabilidad Bruta', color='red')
ax.set_ylabel('Difference', color='green')

ax.grid(True, which='major', axis='y', linestyle='--', linewidth=0.5)
ax.tick_params(axis='x', rotation=45)
fig.tight_layout()
ax.legend()

plt.show()
```
